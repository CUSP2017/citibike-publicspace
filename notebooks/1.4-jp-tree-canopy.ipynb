```python
import geopandas as gpd
import numpy as np
```

```python
df = gpd.read_file('../data/external/tree-canopy.geojson')
```

```python
df.columns.values
```

```python
df.head()
```

```python
keep_columns = ["block_id", "tree_dbh", "status", "health", "spc_latin", 
                "spc_common", "sidewalk", "latitude", "longitude"]

delete_columns = [x for x in df.columns.values if x not in keep_columns]

df.drop(delete_columns, axis=1)
```

```python
# There are dead and stumped trees in the data set!
print ("Dead:", len(df[df.status == "Dead"]))
print ("Stump:", len(df[df.status == "Stump"]))
```

```python
# Keep only trees that are alive 
df = df[df.status == "Alive"]
```
