```python
import pandas as pd
import numpy as np
import os
import csv
import StringIO
import requests
import zipfile
```

```python
datestring = [201501, 201502, 201503, 201504, 201505, 201506, 201507, 201508, 201509, 201510, 201511, 201512]
i = 0
dfs = []
for elem in datestring:
    name = 'citi' + str(datestring[i])
    r = requests.get("https://s3.amazonaws.com/tripdata/" + str(datestring[i]) + "-citibike-tripdata.zip") 
    z = zipfile.ZipFile(StringIO.StringIO(r.content))
    citi_csv = z.extractall("/Users/kristikorsberg/devel/ADS/Citi_Bike_Ridership_Data/")
    name = pd.read_csv("/Users/kristikorsberg/devel/ADS/Citi_Bike_Ridership_Data"+ "/" + str(datestring[i]) + "-citibike-tripdata.csv")
    name = pd.DataFrame(name['tripduration'].groupby(name['start station id']).count())
    tripcount = 'count of trip duration ' + str(datestring[i])
    name = name.rename(columns={'tripduration': tripcount})
    dfs.append(name)
    i+=1
```

```python
total = dfs[0].join((dfs[1], dfs[2], dfs[3], dfs[4], dfs[5], dfs[6], dfs[7], dfs[8], dfs[9], dfs[10], dfs[11]), how='outer')
total['average ridership 2015'] = total.mean(axis=1)
```
